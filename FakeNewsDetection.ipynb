# Preprocessing

```python
def clean_text(text):
    text = text.lower()  # Convert to lowercase
    text = re.sub(r'<.*?>', '', text)  # Remove HTML tags
    text = re.sub(r'https?://\S+|www\.\S+', '', text)  # Remove URLs
    text = re.sub(r'\[.*?\]', '', text)  # Remove text inside brackets
    text = re.sub(r'\S*@\S*\s?', '', text)  # Remove email addresses
    text = re.sub(r'@[A-Za-z0-9_]+', '', text)  # Remove mentions (@username)
    text = re.sub(r'#\S+', '', text)  # Remove hashtags
    text = re.sub(r'\d+', '', text)  # Remove numbers
    text = re.sub(rf"[{re.escape(string.punctuation)}]", '', text)  # Remove punctuation
    text = re.sub(r'\b\w{1,2}\b', '', text)  # Remove short words (1-2 characters)
    text = re.sub(r'\s+', ' ', text).strip()  # Remove extra spaces
    return text
```

```python
data['text'] = data['text'].apply(clean_text)
```

```python
x = data['text']
y = data['class']
```

# Dataset split

```python
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size = 0.25)
```

```python
print(x_train[:5])  # Check if any None values exist
```

```python
# Remove None or NaN values from x_train
x_train = [text if text is not None else "" for text in x_train]  # Replace None with empty string
```

```python
x_train = [str(text) for text in x_train]
```

# Model Training

```python
from sklearn.feature_extraction.text import TfidfVectorizer

vectorization =TfidfVectorizer()
xv_train = vectorization.fit_transform(x_train)
xv_test = vectorization.transform(x_test)
```

# With Logistic Regression

```python
from sklearn.linear_model import LogisticRegression

LR = LogisticRegression()
LR.fit(xv_train, y_train)
```

```python
pred_lr = LR.predict(xv_test)
```

```python
LR.score(xv_test,y_test)
```

```python
print(classification_report(y_test,pred_lr))
```

# With Decision Tree Classifier

```python
from sklearn.tree import DecisionTreeClassifier

DT = DecisionTreeClassifier()
DT.fit(xv_train, y_train)
```

```python
pred_dt = DT.predict(xv_test)
```

```python
DT.score(xv_test,y_test)
```

```python
print(classification_report(y_test,pred_dt))
```

# Testing


```python
# Function to convert predictions to labels
def output_label(n):
    return "Fake News" if n == 0 else "Not Fake News"

# Function to manually test a news article
def manual_testing(news):
    # Preprocess the input news
    testing_news = {"text": [news]}  # Create a dictionary with input text
    new_def_test = pd.DataFrame(testing_news)  # Convert to DataFrame
    new_def_test["text"] = new_def_test["text"].apply(clean_text)  # Apply preprocessing

    # Transform the input text using the trained vectorizer
    new_x_test = new_def_test["text"]
    new_xv_test = vectorization.transform(new_x_test)

    # Make predictions using all four models
    pred_LR = LR.predict(new_xv_test)[0]  # Logistic Regression
    pred_DT = DT.predict(new_xv_test)[0]  # Decision Tree
    pred_GB = GB.predict(new_xv_test)[0]  # Gradient Boosting
    pred_RF = RF.predict(new_xv_test)[0]  # Random Forest
    
    return {
        "Logistic Regression": output_label(pred_LR),
        "Decision Tree": output_label(pred_DT),
        "Gradient Boosting": output_label(pred_GB),
        "Random Forest": output_label(pred_RF)
    }
```

# With Random Internet News

```python
import random

# Function to perform majority voting
def ensemble_prediction(predictions):
    votes = list(predictions.values())  # Get all predictions
    final_prediction = max(set(votes), key=votes.count)  # Get the most frequent prediction
    return final_prediction

# Example usage
news = str(input("Enter the news article: "))  # Textbox input
predictions = manual_testing(news)  # Get predictions from individual models
final_result = ensemble_prediction(predictions)  # Perform majority voting

print("\nPredictions from individual models:")
print(predictions)
print("\nFinal Ensemble Prediction:", final_result)

```
